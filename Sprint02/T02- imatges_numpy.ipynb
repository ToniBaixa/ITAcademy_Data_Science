### Ex4
##### Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Broadcasting d'una matriu 1D
a = np.array([17, 11, 19]) 
print(a)
b = 3 
print(b)
print("Broadcasting d'una matriu 1D") 
c = a + b
print(c)

# Broadcasting d'dos matrius amb la mateixa dimensió
# Same size of the trailing axes for both arrays
a = np.array([[ 0.0,  0.0,  0.0],[10.0, 10.0, 10.0],[20.0, 20.0, 20.0],[30.0, 30.0, 30.0]]) # 4x3
b = np.array([1.0, 2.0, 3.0]) # 4
c= a + b
print("\nBroadcasting d'dos matrius amb la mateixa dimensió") 
print (c)

# Broadcasting d'dos matrius amb dimensions diferents = ERROR
a = np.array([[ 0.0,  0.0,  0.0],[10.0, 10.0, 10.0],[20.0, 20.0, 20.0],[30.0, 30.0, 30.0]]) # 4x3
b = np.array([0.0, 1.0, 2.0, 3.0]) # 4
print("\nBroadcasting de dues matrius amb dimensions diferents = ERROR") 
c= a + b
print (c)
```

### Ex5
##### Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
matriu = np.array([
	[21, 22, 23],
	[34, 35, 36],
	[47, 48, 49],
	[55, 56, 57]
])
num= 1 # triem un número, possible
colum = matriu[:,num]
print("Columna {}= {}". format(num,colum))
row = matriu[num]
print("Fila {}= {}". format(num, row))
print("\nBroadcasting de aquestes dues matrius que tenen dimensions diferents = ERROR") 
print("Columna + Fila = {}". format( colum+row))
```
