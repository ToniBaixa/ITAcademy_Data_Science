### Ex1
##### Agafa un conjunt de dades de tema esportiu que t'agradi. Realitza un mostreig de les dades generant una mostra aleatòria simple i una mostra sistemàtica.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import random


df = pd.read_csv("Stats.csv")

df.head()
```

```python
df = df.reset_index()
df.head()
```

##### mostra aleatòria simple

```python
simple_random_sampling = df.sample(100)
print (simple_random_sampling)
```

##### una mostra sistemàtica.

```python
sample = 1000
interval =  int (df.size/ sample)
random_number = random.randint(1, 30)

# Define systematic sampling function
def systematic_sampling(df, step):
 
    indexes = np.arange(random_number, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
 
# Obtain a systematic sample and save it in a new variable
systematic_sample = systematic_sampling(df, interval)
 
# View sampled data frame
systematic_sample.head(5)
```

##### genera una mostra estratificada 

```python
df['winner'].value_counts()
```

```python
#So 50% winners & 50% losers. It's a tennis game!! :-)
# .apply(lambda x: x.sample(N))
N= 20
df.groupby('winner', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(df))))).sample(frac=1).reset_index(drop=True)
```
